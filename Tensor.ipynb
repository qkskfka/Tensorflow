```python
import tensorflow as tf
```

```python
# 변수 a와 b를 미리 정의합니다.
a = tf.Variable(0.1)
b = tf.Variable(0.2)

# 키와 신발 사이의 관계를 표현하는 식입니다.
def 신발_예측(키):
    return 키 * a + b

# 손실 함수를 정의합니다.
def 손실함수(키):
    예측값 = 신발_예측(키)
    return tf.square(260 - 예측값)

# tf.function 데코레이터를 사용하여 손실 함수를 컴파일합니다.
@tf.function
def 최적화_단계(키):
    with tf.GradientTape() as tape:
        손실 = 손실함수(키)
    gradients = tape.gradient(손실, [a, b])
    opt.apply_gradients(zip(gradients, [a, b]))

# 옵티마이저를 정의합니다.
opt = tf.keras.optimizers.Adam(learning_rate=0.1)

# 옵티마이저를 사용하여 손실 함수를 최소화합니다.
for i in range(300):
    최적화_단계(170)  # 키 값을 직접 지정하여 사용
    print(a.numpy(), b.numpy())
```
